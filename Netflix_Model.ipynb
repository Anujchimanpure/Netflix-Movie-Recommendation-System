```python
# finds similarity

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Clean descriptions
df['description'] = df['description'].fillna('').astype(str)

# Build TF-IDF matrix
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(df['description'])

# Compute similarity
similar = cosine_similarity(tfidf_matrix)

print(similar.shape)
```

```python
def recommend_movie(title):
    title = title.lower().strip()

    df['title_clean'] = df['title'].str.lower().str.strip()

    if title not in df['title_clean'].values:
        print("Movie not found in dataset")
        return

    idx = df[df['title_clean'] == title].index[0]

    similar_score = list(enumerate(similar[idx]))
    similar_score = sorted(similar_score, key=lambda x: x[1], reverse=True)

    print("Recommended Movies:")
    for i in similar_score[1:6]:
        print(df.iloc[i[0]]['title'])
```

```python
recommend_movie('Kota Factory')
```

```python
recommend_movie('inception')
```

```python
import pandas as pd
import pickle

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv("netflix_clean_data.csv")

# Use only needed columns
df = df[['title', 'description']]

# Handle missing values
df['description'] = df['description'].fillna('')
```

```python
tfidf = TfidfVectorizer(stop_words='english')

tfidf_matrix = tfidf.fit_transform(df['description'])
```

```python
similarity = cosine_similarity(tfidf_matrix)
```

```python
with open("similarity.pkl", "wb") as f:
    pickle.dump(similarity, f)
```
